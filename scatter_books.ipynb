# Plot to ord mot hverandre

Her er det pandas scatter-plot i aksjon for å se på fordeling av ordpar i tekster

```python
import pandas as pd
import dhlab.text as dh
```

```python
def scatter_plot(df, x = ".", y = ",", **kwargs):
    return df.loc[[x,y]].transpose().plot.scatter(x, y, **kwargs)
```

Definer et korpus. Artig å sammenligne forskjellige. Ved bruk av telling er det en øvre grense på tusen bøker. Det kan ta tid å laste ned alt, men oppgaven kan gjøres i bolker om nødvendig.

```python
corpus = dh.Corpus(doctype = "digibok", from_year = 1990, to_year=2020, limit = 1000, subject = "kvinnesak")
corpus.size
```

```python
mycounts = dh.Counts(corpus = corpus, words = ". , og i han hun hit dit høy lav".split())
```

Ta opptellingen over i variabelen `df` og normaliser samtidig. 

```python
df = (mycounts.counts/mycounts.counts.sum()).fillna(0)
```

Starter med punktum og komma. Komma-aksen er samme som y-akse, og punktum-aksen er x-akse.

```python
ax = scatter_plot(df, x = '.', y = ',',c = 'blue', alpha = 0.7)
scatter_plot(df, x = "og", y = "i", c = 'red', ax = ax, alpha = 0.7)
scatter_plot(df, x = "han", y = "hun", c = 'green', ax = ax, alpha = 0.7);
```

Plot mellom _og_ og _i_.

```python
scatter_plot(df, x = "og", y = "i", ax = ax)
```

Forholdet mellom _han_ og _hun_, legg merke til at plottet er tett ned langs _han_-aksen. Selv der det er mye _hun_ er det fler _han_.

```python
scatter_plot(df, x = 'han', y = "hun")
```
